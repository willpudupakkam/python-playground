```python
import random as rand
import matplotlib.pyplot as plt
import numpy as np
import time

def galton(l, b):
    """Plot a histogram of final positions of 'b' balls on a Galton Board with 'l' levels"""

    data = []
    
    n = 0
    while n < b:
        data.append(sum(map(lambda x : 0.5 if x > 0.5 else -0.5, [rand.random() for i in np.zeros(l)])))
        n += 1
    
    plt.hist(np.array(data), bins=l+1, range=(-0.5*(l+1), 0.5*(l+1)))
    plt.show()

t0 = time.time()
galton(10,10000)
t1 = time.time()
print("Evaluated in {} seconds".format(t1-t0)) 
```
